```
rawDF= spark.read.option("delimiter",";").option("header",True).option("inferSchema", True).csv("/FileStore/tables/marketds4spark-1.csv")


#inferSchema true dediğimiz zaman veri tiplerini kendi veri tipine dönüştürür.
```

```
rawDF.printSchema()  #kolon tipi için yani kolon adlarının veri tipini gösteririr
```

```
rawDF.filter("ITEMNAME= 'KEREVIZ'").show()
```

```
sutKahvaltıDF = rawDF.filter("CATEGORY_NAME1 = 'SÜT KAHVALTILIK'")
```

```
sutKahvaltıDF.show()
```

```
veganDF = rawDF.filter("CATEGORY_NAME1 = 'İÇECEK' or CATEGORY_NAME1 ='MEYVE SEBZE' or CATEGORY_NAME1 ='KAĞIT'")
```

```
veganDF.show()
```

```
veganDF = rawDF.filter(("CATEGORY_NAME1 = 'İÇECEK' or CATEGORY_NAME1 ='MEYVE SEBZE' or CATEGORY_NAME1 ='KAĞIT'") and  " GENDER= 'E'" ) #içecek veya meyve alan erkekler 
```

```
veganDF.show()
```

```
nameItem = veganDF.select('CLIENTCODE','ITEMNAME','CITY') #select kolonlar filtrelenir filter ise satırları filtreler
```

```
nameItem.show()
```

```
rawDF.groupBy('CITY').count().show()
```

```
countByCity = rawDF.groupBy('CITY').count()
```

```
countByCity.show()
```

```
countByCity.orderBy('count' , ascending = False).show()   #en küçükten büyüğe , ascending kapatılarak büyükten küçüğe yaptık 
```

```
rawDF.groupBy('City').sum('LINENETTOTAL').show() #işlem hacmi
```

```
rawDF.groupBy('City').sum('LINENETTOTAL').withColumnRenamed('sum(LINENETTOTAL)' ,'totalValue').show()  #kolon ismini düzenledik
```

```
rawDF.groupBy('City').sum('LINENETTOTAL').withColumnRenamed('sum(LINENETTOTAL)' ,'totalValue').orderBy('totalValue' , ascending = False).show()  #büyükten küçüğe 
```

```
rawDF.withColumn('customerType', when(col("LINENETTOTAL")>20 ,"ELITE PLUS"))  #alışveriş tutarından müşteri ayıkladık
```

```
rawDF.withColumn('customerType', when(col("LINENETTOTAL")>20 ,"ELITE PLUS")).filter("customerType = 'ELITE PLUS'")
```

```
rawDF.withColumn('customerType', when(col("LINENETTOTAL")>20 ,"ELITE PLUS")).filter("customerType = 'ELITE PLUS'").show() #elite plus olanları getirdik filtreleyerek.
```

```
rawDF.withColumn('customerType', when(col("LINENETTOTAL")>20 ,"ELITE PLUS").when(col("LINENETTOTAL")<20 ,"STANDART")).show()  #20 den küçük olanlara standart değer ekledik
```

```
rawDF.withColumn('customerType', when(col("LINENETTOTAL")>20 ,"ELITE PLUS")
                 .when((col("LINENETTOTAL")<20) & (col("LINENETTOTAL")<10) ,"STANDART")).show() #20 ve 10 dan küçük olanlara
```
